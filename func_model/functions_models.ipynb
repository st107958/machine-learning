```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
```

```python
X = np.linspace(0, 10, 100)
y = np.exp(X) 
```

```python
X = np.linspace(0, 10, 100)
y = np.cos(X) 
```

```python
X = np.linspace(1, 10, 100)
y = np.log(X) 
```

```python
X = np.linspace(-10, 10, 100)
y = np.power(X, 5) 
```

```python
X = np.linspace(0, 10, 100)
y = np.power(3, X) 
```

```python
X = X.reshape(-1, 1)

poly_features = PolynomialFeatures(degree=6)
poly_model = make_pipeline(poly_features, LinearRegression())
poly_model.fit(X, y)
poly_predictions = poly_model.predict(X)

poly_mse = mean_squared_error(y, poly_predictions)
print(f'MSE: {poly_mse}')

plt.scatter(X, y, color='blue', label='Function', s=1)
plt.plot(X, poly_predictions, color='red', label='Polynomial Regression')
plt.legend()
plt.show()
```
