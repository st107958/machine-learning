```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load('X_overlapping.npy')
y = np.load('y_overlapping.npy')
n_clusters=3
```

```python
X = np.load('X_circles.npy')
y = np.load('y_circles.npy')
n_clusters=2
```

```python
X = np.load('X_moons.npy')
y = np.load('y_moons.npy')
n_clusters=2
```

```python
X = np.load('X_chessboard.npy')
y = np.load('y_chessboard.npy')
n_clusters=2
```

```python
r = np.sqrt(X[:,0]**2 + X[:,1]**2)
theta = np.arctan2(X[:,1], X[:,0])

X = np.column_stack((r, theta))
```

```python
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
clustering = AgglomerativeClustering(n_clusters=n_clusters)
clustering.fit(X)
labels = clustering.labels_
```

```python
X[:,0] = r * np.cos(theta)
X[:,1] = r * np.sin(theta)
```

```python
fig, axs = plt.subplots(1, 2, layout='constrained')
axs[0].scatter(X[:, 0], X[:, 1], c=y)
axs[0].set_title('Original data')
axs[1].scatter(X[:, 0], X[:, 1], c=labels)
axs[1].set_title('Clustered data')
plt.show()
```
